```python
from landfall import plot_points
```

```python
# Define points data
lats = [27.856968520151316,
        27.911342150381646,
        27.94362362728143,
        27.95332511902476]
lons = [-82.49279919372195,
       -82.49663445374274,
       -82.47890194094782,
       -82.45522287566286]
colors = ['#ba7ba7',
          '#b33951',
          '#f58a07',
          '#f58a07']
ids = [1, 1, 2, 3]
```

```python
# Just plot points, all same color
plot_points(lats, lons)
```

```python
# plot points with specified colors
plot_points(lats, lons, colors=colors)
```

```python
# plot points with each getting a random color
plot_points(lats, lons, colors='random')
```

```python
# plot points with each getting an evenly spread color wheel color
plot_points(lats, lons, colors='wheel')
```

```python
# plot points with each getting a visually distict color
plot_points(lats, lons, colors='distinct')
```

```python
# plot points with each unique id passed (ids) getting an assigned color (id_colors)
plot_points(lats, lons,
            ids=ids,
            id_colors={1: '#FF007F', 2: '#00CC99', 3: '#8A2BE2'})
```

```python
# plot points with each unique id passed (ids) getting a random color
plot_points(lats, lons, ids=ids, id_colors='random')
```

```python
# plot points 
plot_points(lats, lons, ids=ids, id_colors='wheel')
```

```python
plot_points(lats, lons, ids=ids, id_colors='distinct')
```
